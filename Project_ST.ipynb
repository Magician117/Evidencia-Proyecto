**Color Palette detector**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from sklearn.cluster import KMeans


# Image into pixels
image = cv2.imread('imagen1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
pixels = image.reshape(-1, 3)

# Get the 4 most dominant colors
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(pixels)
colors = np.array(kmeans.cluster_centers_, dtype='uint8')

# Dominant Colors
plt.figure(figsize=(8, 4))
plt.subplot(1, 4, 1)
plt.imshow([[colors[0]/255]])
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow([[colors[1]/255]])
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow([[colors[2]/255]])
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow([[colors[3]/255]])
plt.axis('off')
```
